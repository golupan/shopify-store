```python
import pandas as pd

google_sheet_url = "https://docs.google.com/spreadsheets/d/19zrekAnhJkQrXoR4Aoj665FoXJ4lmLlrAl-d47SbK6c/export?format=csv"
df_raw = pd.read_csv(google_sheet_url)

display(df_raw.head())
display(df_raw.info())
```

```python
df_raw['Timestamp'] = pd.to_datetime(df_raw['Timestamp'], errors='coerce')
display(df_raw.head())
display(df_raw.info())
```

```python
print("Unique 'Product Category' values:")
print(df_raw['Product Category'].unique())
print("\nUnique 'Fulfillment Status' values:")
print(df_raw['Fulfillment Status'].unique())
```

```python
product_category_mapping = {
    'Cocktail Dreses': 'Cocktail Dresses',
    'PromDrsses': 'Prom Dresses'
}
df_raw['Product Category'] = df_raw['Product Category'].replace(product_category_mapping)

fulfillment_status_mapping = {
    'Delaye': 'Delayed',
    'Fufilled': 'Fulfilled'
}
df_raw['Fulfillment Status'] = df_raw['Fulfillment Status'].replace(fulfillment_status_mapping)
```

```python
print("Unique 'Product Category' values after standardization:")
print(df_raw['Product Category'].unique())
print("\nUnique 'Fulfillment Status' values after standardization:")
print(df_raw['Fulfillment Status'].unique())
```

```python
columns_to_fill = ['Rating', 'Review Content', 'Order Value', 'Product Category', 'Fulfillment Status']

df_cleaned = df_raw.copy()

df_cleaned['Rating'] = df_cleaned['Rating'].fillna('Unknown')
df_cleaned['Review Content'] = df_cleaned['Review Content'].fillna('')
df_cleaned['Order Value'] = df_cleaned['Order Value'].fillna('Unknown')
df_cleaned['Product Category'] = df_cleaned['Product Category'].fillna('Unknown')
df_cleaned['Fulfillment Status'] = df_cleaned['Fulfillment Status'].fillna('Unknown')

display(df_cleaned[columns_to_fill].isnull().sum())
```

```python
df_filtered = df_cleaned[df_cleaned['Review Content'].str.len() > 10]
df_filtered = df_filtered[df_filtered['Rating'] != 'Unknown']
display(df_filtered.head())
```

```python
df_filtered.to_csv('cleaned_reviews.csv', index=False)
```

```python
markdown_table = df_filtered.to_markdown(index=False)
print(markdown_table)
```

```python
base_prompt_template = """
Analyze the following customer review data and answer the questions below.

Cleaned Data:
{cleaned_data}

Questions:
1. Which product categories have the most 1-star reviews in Canada?
2. Do higher order values correlate with lower ratings?
3. Can you summarize the top 5 complaints and top 5 compliments across all reviews?
4. Which fulfillment statuses are most associated with negative feedback?
"""

ai_prompt = base_prompt_template.format(cleaned_data=markdown_table)

print(ai_prompt)
```

```python
ai_answers = {
    "most_1_star_in_canada": "Based on the provided data, 'Prom Dresses' and 'Unknown' product categories have the most 1-star reviews in Canada.",
    "order_value_vs_rating": "The data does not show a clear correlation between higher order values and lower ratings. The relationship appears to be inconsistent across the dataset.",
    "top_complaints_and_compliments": {
        "complaints": [
            "Issues with product quality or receiving the wrong item.",
            "Delays in shipping and fulfillment.",
            "Poor customer service experience.",
            "Sizing and fit problems with clothing.",
            "Difficulty with the returns process."
        ],
        "compliments": [
            "Positive feedback on the design and style of the dresses.",
            "Appreciation for fast shipping and timely delivery.",
            "Good quality and material of the products.",
            "Excellent customer service and support.",
            "Accurate sizing and fit of the garments."
        ]
    },
    "fulfillment_status_and_feedback": "The 'Cancelled' and 'Delayed' fulfillment statuses are most associated with negative feedback, as indicated by lower average ratings for these statuses."
}

import json
print(json.dumps(ai_answers, indent=2))
```

```python
def get_ai_answer(question):
    """
    Retrieves the answer for a given question from the ai_answers dictionary.

    Args:
        question: The question string.

    Returns:
        The corresponding answer if found, otherwise a default message.
    """
    # Normalize the input question for easier matching (case-insensitive, remove punctuation)
    normalized_question = question.lower().strip()

    # Check for matches with predefined questions
    if "most 1-star reviews in canada" in normalized_question:
        return ai_answers.get("most_1_star_in_canada", "Answer for 'most 1-star reviews in Canada' not available.")
    elif "higher order values correlate with lower ratings" in normalized_question or "order value vs rating" in normalized_question:
        return ai_answers.get("order_value_vs_rating", "Answer for 'order value vs rating' not available.")
    elif "top 5 complaints and top 5 compliments" in normalized_question or "complaints and compliments" in normalized_question:
        complaints = ai_answers.get("top_complaints_and_compliments", {}).get("complaints", [])
        compliments = ai_answers.get("top_complaints_and_compliments", {}).get("compliments", [])
        if complaints or compliments:
            response = "Top Complaints:\n" + "\n".join([f"- {c}" for c in complaints]) + "\n\nTop Compliments:\n" + "\n".join([f"- {c}" for c in compliments])
            return response
        else:
            return "Answer for 'top complaints and compliments' not available."
    elif "fulfillment statuses most associated with negative feedback" in normalized_question or "fulfillment status and feedback" in normalized_question:
        return ai_answers.get("fulfillment_status_and_feedback", "Answer for 'fulfillment status and feedback' not available.")
    else:
        return "I don't have a specific answer for that question. Please try rephrasing or ask about: 1-star reviews in Canada, order value vs rating, top complaints and compliments, or fulfillment status and feedback."

print("AI Assistant: Hello! Ask me a question about the customer reviews (e.g., 'Which product categories have the most 1-star reviews in Canada?'). Type 'quit' to exit.")

while True:
    user_input = input("You: ")
    if user_input.lower() == 'quit':
        break
    response = get_ai_answer(user_input)
    print(f"AI Assistant: {response}")
```
